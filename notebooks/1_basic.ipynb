```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
          
```

#### static prompt


```python
prompt=[
    ("system","you are devops enginner"),
    ("user","how to create docker file")
]
res=llm.invoke(prompt)
```

```python
prompts=ChatPromptTemplate.from_messages([
    {"role":"system","content":"you are a {domain} engineer"},
    {"role":"user","content":"{query}"},
])
```

```python
chains=  prompts | llm
res=chains.invoke({"domain":"devops" , "query":"steps to create docker image"})
```

#### basic memory


```python
history=[]
while True:
    query=input("user")
    if query.lower() in ["exit", "bye","stop"]:
        print("good Bye")
        break
    
    history.append({"role":"user", "content":query})
    print(query)

    res=llm.invoke(history)
    history.append({"role":"ai","content":res.content})
    print(res.content )  


```
