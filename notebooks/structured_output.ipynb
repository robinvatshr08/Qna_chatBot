```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_google_genai import ChatGoogleGenerativeAI
from pydantic import BaseModel, Field
llm=GoogleGenerativeAI(model="gemini-2.5-flash")
str_llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

### discuss problem

```python
text="my name is robin.my age is 28 and i am from kaithal"
res=llm.invoke(f"give me only name and age from {text}")
```

```python
print(res)
```

### structured output

```python
class ResponseStrecture(BaseModel):
    name: str = Field(description="complete name")
    age:int = Field(description="age")
    email:str= Field(description="email")

str_llm2=str_llm.with_structured_output(ResponseStrecture)
```

```python
res=str_llm2.invoke("my name is robin.my age is 28 and my email is awe@gmail.com")
```

```python
res.name
res.age
res.email
res.model_dump()
```
